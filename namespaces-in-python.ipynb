### Namespaces:
#### LEGB
#### Local, Enclosing, Global Built-in
* Local: Variables defined within a function
* Enclosing: Variables in the local scope of enclosing function 
* Global: Variables defined at top level of module or explicitly declared global 
* Built-in: names prea ssigned in python

#### Python firsts checks variables in local scope, then enclosing, then global and then built  in that's why abbreviation LEGB

!

```python
x='global'
def test():
    y='local y'
    print(y) #python checks first is y in local scope
test()
```

```python
x='global x'
def test():
    y='local y'
    print(x) #python checks first is x in local scope
    # if not then checks in enclosing scope.
    # if not then checks in global scope. 
test()
```

```python
x='global x'
def test():
    y='local y'
    print(x) #python checks first is x in local scope
    # if not then checks in enclosing scope.
    # if not then checks in global scope. 
test()
print(y) ## error because y not present in local, enclosed, global and built-in scope
```

```python
x='global x'
def test():
    x='local x' #different x
    print(x) #python checks first is x in local scope
    
test()
print(x)
```

```python
x='global x'
def test():
    global x #here we explicitly tell python that use global x
    x='local x' #resetting global x
    print(x) #python checks first is x in local scope
    
test()
print(x)
```

```python
# x='global x' # we don't need to define here x if we're setting x as global variable inside the function
def test():
    global x #here we explicitly tell python that use global x
    x='local x' #resetting global x
    print(x) #python checks first is x in local scope
    
test()
print(x)
```

#### Enclosing Scope
Nested functions

```python
def outer():
    x='outer x'#local to outer function
    def inner():
        x='inner x' #local to inner function
        print(x)
    inner()
    print(x)
    
outer()
    
```

```python
def outer():
    x='outer x'#local to outer function
    def inner():
        # x='inner x' 
        print(x) # checks any x in local function no,
        # then checks if x present in local scope of enclosing function
        # it finds yes x  is there in local scope of enclosing function
    inner()
    print(x)
    
outer()
```
